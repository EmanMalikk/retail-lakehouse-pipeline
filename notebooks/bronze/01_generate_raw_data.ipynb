```python
# Install Faker if needed
# Run this once if Faker gives import error:
# %pip install faker

from faker import Faker
import pandas as pd
import random
from datetime import datetime
import os

fake = Faker()

# -----------------------------
# CONFIG
# -----------------------------
NUM_CUSTOMERS = 100
NUM_PRODUCTS = 50
NUM_ORDERS = 200

# -----------------------------
# CUSTOMERS
# -----------------------------
customers = []

for customer_id in range(1, NUM_CUSTOMERS + 1):
    customers.append({
        "customer_id": customer_id,
        "customer_name": fake.name(),
        "email": fake.email(),
        "city": fake.city(),
        "country": fake.country(),
        "signup_date": fake.date_between(start_date='-2y', end_date='today')
    })

customers_df = pd.DataFrame(customers)

# -----------------------------
# PRODUCTS
# -----------------------------
categories = ["Electronics", "Clothing", "Home", "Sports", "Beauty"]

products = []

for product_id in range(1, NUM_PRODUCTS + 1):
    products.append({
        "product_id": product_id,
        "product_name": fake.word().capitalize(),
        "category": random.choice(categories),
        "price": round(random.uniform(10, 1000), 2),
        "stock_quantity": random.randint(10, 500)
    })

products_df = pd.DataFrame(products)

# -----------------------------
# ORDERS
# -----------------------------
orders = []

for order_id in range(1, NUM_ORDERS + 1):
    customer_id = random.randint(1, NUM_CUSTOMERS)

    orders.append({
        "order_id": order_id,
        "customer_id": customer_id,
        "order_date": fake.date_between(start_date='-1y', end_date='today'),
        "order_status": random.choice([
            "Pending",
            "Shipped",
            "Delivered",
            "Cancelled"
        ])
    })

orders_df = pd.DataFrame(orders)

# -----------------------------
# ORDER ITEMS
# -----------------------------
order_items = []

for item_id in range(1, NUM_ORDERS * 3):
    order_items.append({
        "order_item_id": item_id,
        "order_id": random.randint(1, NUM_ORDERS),
        "product_id": random.randint(1, NUM_PRODUCTS),
        "quantity": random.randint(1, 5)
    })

order_items_df = pd.DataFrame(order_items)

# -----------------------------
# PAYMENTS
# -----------------------------
payments = []

for payment_id in range(1, NUM_ORDERS + 1):
    payments.append({
        "payment_id": payment_id,
        "order_id": payment_id,
        "payment_method": random.choice([
            "Credit Card",
            "Debit Card",
            "PayPal",
            "Cash"
        ]),
        "payment_status": random.choice([
            "Paid",
            "Failed",
            "Refunded"
        ])
    })

payments_df = pd.DataFrame(payments)

# -----------------------------
# SHIPPING
# -----------------------------
shipping = []

for shipping_id in range(1, NUM_ORDERS + 1):
    shipping.append({
        "shipping_id": shipping_id,
        "order_id": shipping_id,
        "shipping_provider": random.choice([
            "DHL",
            "FedEx",
            "UPS"
        ]),
        "shipping_status": random.choice([
            "In Transit",
            "Delivered",
            "Delayed"
        ])
    })

shipping_df = pd.DataFrame(shipping)

# -----------------------------
# SAVE CSV FILES
# -----------------------------

base_path = "/Workspace/Repos/emaanadeem0@gmail.com/E-com-Project/data/raw/"

# Create directory if not exists
os.makedirs(base_path, exist_ok=True)

# Save files
customers_df.to_csv(base_path + "customers.csv", index=False)
products_df.to_csv(base_path + "products.csv", index=False)
orders_df.to_csv(base_path + "orders.csv", index=False)
order_items_df.to_csv(base_path + "order_items.csv", index=False)
payments_df.to_csv(base_path + "payments.csv", index=False)
shipping_df.to_csv(base_path + "shipping.csv", index=False)

print("Raw CSV files generated successfully!")
```
