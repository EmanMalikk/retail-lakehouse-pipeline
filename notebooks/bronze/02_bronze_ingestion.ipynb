```python
# -----------------------------
# BASE PATH
# -----------------------------

base_path = "/Workspace/Repos/emaanadeem0@gmail.com/E-com-Project/data/raw/"

# -----------------------------
# READ CSV FILES
# -----------------------------

customers_df = spark.read.format("csv") \
    .option("header", "true") \
    .option("inferSchema", "true") \
    .load(base_path + "customers.csv")

products_df = spark.read.format("csv") \
    .option("header", "true") \
    .option("inferSchema", "true") \
    .load(base_path + "products.csv")

orders_df = spark.read.format("csv") \
    .option("header", "true") \
    .option("inferSchema", "true") \
    .load(base_path + "orders.csv")

order_items_df = spark.read.format("csv") \
    .option("header", "true") \
    .option("inferSchema", "true") \
    .load(base_path + "order_items.csv")

payments_df = spark.read.format("csv") \
    .option("header", "true") \
    .option("inferSchema", "true") \
    .load(base_path + "payments.csv")

shipping_df = spark.read.format("csv") \
    .option("header", "true") \
    .option("inferSchema", "true") \
    .load(base_path + "shipping.csv")

print("CSV files loaded successfully!")
```

```python
customers_df.show(5)
```

```python
orders_df.show(5)
```

```python
spark.sql("CREATE DATABASE IF NOT EXISTS bronze")
```

```python
# -----------------------------
# WRITE DELTA TABLES
# -----------------------------

customers_df.write.format("delta") \
    .mode("overwrite") \
    .saveAsTable("bronze.bronze_customers")

products_df.write.format("delta") \
    .mode("overwrite") \
    .saveAsTable("bronze.bronze_products")

orders_df.write.format("delta") \
    .mode("overwrite") \
    .saveAsTable("bronze.bronze_orders")

order_items_df.write.format("delta") \
    .mode("overwrite") \
    .saveAsTable("bronze.bronze_order_items")

payments_df.write.format("delta") \
    .mode("overwrite") \
    .saveAsTable("bronze.bronze_payments")

shipping_df.write.format("delta") \
    .mode("overwrite") \
    .saveAsTable("bronze.bronze_shipping")

print("Bronze Delta tables created successfully!")
```
